## Plotting BacGen project

#### Import packages

```python
# pandas for work with tables
import pandas as pd

# seaborn for nice higher level plotting
import seaborn as sns

# matplotlib for base level plotting
import matplotlib.pyplot as plt 
```

#### Import table

```python
df = pd.read_csv("data.txt", sep="|", header=None)
```

```python
d = df[0].str.split(" ", expand=True, n=2)
d
```

```python
d.columns = ["gc", "genes", "phylum"]
```

```python
# define variable types to allow for plotting (default is string)

d.gc = d.gc.astype(float)
d.genes = d.genes.astype(int)
```

```python
d.phylum
```

#### Plotting

```python
g = sns.relplot(data=d, x="genes" , y="gc", hue="phylum", legend=False, aspect=1.5, alpha=.4, col="phylum", height=1.5, col_wrap=7)
g.set_titles("{col_name}")
```
